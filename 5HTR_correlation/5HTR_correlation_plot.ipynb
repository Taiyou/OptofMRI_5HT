```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import scipy.io
```

```python
df1 = pd.read_csv('htRmatrix.csv');
df2 = pd.read_csv('htPmatrix.csv');
```

```python
df1.head()
```

```python
df2.head()
```

```python
np.zeros_like(df1)
```

```python
mask = np.zeros_like(df1, dtype=np.bool)
```

```python
mask[np.triu_indices_from(mask)]= True
```

```python
sns.set_context('paper')
f, ax = plt.subplots(figsize=(12, 17)) 
heatmap = sns.heatmap(df1, 
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap ='coolwarm',
                      cbar_kws={"shrink": 0.4, 'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -0.5, 
                      vmax = 0.8,
                      annot = True,
                      annot_kws = {"size": 18})
ax.set_yticklabels(df1.columns, rotation = 0, fontdict = {'fontsize':18})
ax.set_xticklabels(df1.columns, fontdict = {'fontsize':18}) #add the column names as labels
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
```

```python
f, ax = plt.subplots(figsize=(12, 17)) 
heatmap = sns.heatmap(df2, 
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap ='viridis_r',
                      cbar_kws={"shrink": .4, 'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = 0, 
                      vmax = 0.05,
                      annot = True,
                      annot_kws = {"size": 18})
ax.set_yticklabels(df2.columns, rotation = 0,  fontdict = {'fontsize':18})
ax.set_xticklabels(df2.columns,  fontdict = {'fontsize':18}) #add the column names as labels
sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
```
